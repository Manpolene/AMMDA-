Setting the limits of the colorbar

In instances where multiple plots share a similar data range but may have slightly different values, the automatically generated colorbars can vary. To ensure a more accurate comparison between these plots, it is beneficial to set fixed colorbar limits using `vmin` (value minimum) and `vmax` (value maximum). By doing so, not only can the colorbar limits be adjusted to more visually appealing values, but consistency can be maintained across multiple plots, facilitating a clearer understanding of the data.

**Access subplot**

Until now we only created empty subplots, to be more precise we created the axes (the plotting area) for both subplots using the function subplots. \
You see that two variables are assigned to the output of this fucntion, fig and axs. fig is like the figure we earlier creted using plt.figure, wile axs holds the four subplot axes:

**axs**

You see it is just an array containing for subplot axes, and like a normal array before you can acces single elements by indexing: \
To plot now in this axes we select the axes to plot in by axs[i,j] and add a .plot(x,y) to plot our x and y data into this specific axes:

```python
fig, axs = plt.subplots(2,2,sharey=True,sharex=True,figsize=(8,8))

x = np.arange(0,10,0.1)
y = np.sin(x)

axs[0,0].plot(x,y)
axs[1,1].plot(x,y**2)
axs[0,1].plot(x,2*y,color='C1')

plt.show()
plt.close()
```

**Subplot for two y-axis**

Subplots can not only be used to create multiple individual plotting areas, they can also be used to create own plot with two seperate y axis

```python
#1. Create a 1,1 subplot
fig, ax1 = plt.subplots(1,1,figsize=(10,10))

x = np.arange(0,10,0.1)
y = 5*x + 2
y2 = np.exp(x)

ax1.plot(x,y)

plt.show()
plt.close()
```

1. Create a 1,1 subplot
2. create a new axes, ax2, based on ax1 with a twin x-axis
3. plot in this new axes

```python
#1. Create a 1,1 subplot
fig, ax1 = plt.subplots(1,1,figsize=(10,10))

x = np.arange(0,10,0.1)
y = 5*x + 2
y2 = np.exp(x)

ax1.plot(x,y)

ax2 = ax1.twinx() #creates a new axes, ax2, based on ax1 with a twin x-axis

ax2.plot(x,y2) #plot in this new axes

plt.show()
plt.close()
```

And change the color of the second curve to make the distinguishable


```python
#1. Create a 1,1 subplot
fig, ax1 = plt.subplots(1,1,figsize=(10,10))

x = np.arange(0,10,0.1)
y = 5*x + 2
y2 = np.exp(x)

ax1.plot(x,y)

ax2 = ax1.twinx() #creta a new axes, ax2, based on ax1 with a twin x-axis

ax2.plot(x,y2,color='red') #plot in this new axes

plt.show()
plt.close()
```

since ax2 (and also ax1) now is just a normal axes like the single subplots before, you can use all the functions above to edit the plot

```python
#1. Create a 1,1 subplot
fig, ax1 = plt.subplots(1,1,figsize=(10,10))

x = np.arange(0,10,0.1)
y = 5*x + 2
y2 = np.exp(x)

ax1.plot(x,y)
ax1.set_ylabel('y',size=15)

ax2 = ax1.twinx() #creta a new axes, ax2, based on ax1 with a twin x-axis

ax2.plot(x,y2,color='red') #plot in this new axes
ax2.set_ylabel('y2',size=20)

plt.show()
plt.close()
```

To make the two axes more distinguishable, and easy to see which axis belongs to which curve, you can also color the axis ticks with an argument labelcolor in the ticks_params from above:

```python
#1. Create a 1,1 subplot
fig, ax1 = plt.subplots(1,1,figsize=(10,10))

x = np.arange(0,10,0.1)
y = 5*x + 2
y2 = np.exp(x)

ax1.plot(x,y)
ax1.set_ylabel('y',size=15)

#2. create a new axes, ax2, based on ax1 with a twin x-axis
ax2 = ax1.twinx() 

ax2.plot(x,y2,color='red') #plot in this new axes
ax2.set_ylabel('y2',size=20, color='red')
ax2.tick_params(axis='y', which='major', labelsize=15, labelcolor='red')

plt.show()
plt.close()
```
