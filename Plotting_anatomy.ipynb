```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
```

```python
np.random.randn(19680801)

x = np.linspace(0.5, 3.5, 100)
y1 = 3 + np.cos(x)
y2 = 1 + np.cos(1 + x / 0.75) / 2
y3 = np.random.uniform(y1, y2, len(x))
```

## Figure size

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1,1,1, aspect = 1)

plt.show()
plt.close()
```

## Color, linewidth, zorder

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)

ax.plot(x, y1, c=(0.25,0.25,1.00), lw = 2, label = 'Blue signal', zorder =10)

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c='blue', lw=2, label="Blue signal", zorder=10)


plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c='yellow', lw=2, label="Blue signal", zorder=10)


plt.show()
plt.close()
```

## Zorder

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c='blue', lw=2, label="Blue signal", zorder=10)
ax.plot(x, y2, c='red', lw=2, label="Red signal")
ax.plot(x, y1, c='green', lw=15, label="Blue signal", zorder=5)
ax.plot(x, y2, c='red', lw=2, label="Red signal")
ax.scatter(x,y3)


plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c='blue', lw=2, label="Blue signal", zorder=10)
ax.plot(x, y2, c='red', lw=2, label="Red signal")
ax.plot(x, y1, c='green', lw=15, label="Blue signal", zorder=15)
ax.plot(x, y2, c='red', lw=2, label="Red signal")
ax.scatter(x,y3)


plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c='blue', lw=2, label="Blue signal", zorder=10)
ax.plot(x, y2, c='red', lw=2, label="Red signal")
ax.plot(x, y1, c='green', lw=15, label="Blue signal", zorder=15)
ax.plot(x, y2, c='red', lw=2, label="Red signal")
ax.scatter(x,y3, s=200, c='Black')


plt.show()
plt.close()
```

## Building up a plot

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)

ax.plot(x, y1, c='blue', lw=2, label='Blue signal', zorder = 10)
ax.plot(x,y2,  c='red', lw=2, label='Red signal')

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10) #Note: you see here the color defined as RBG, you can also name them as before
ax.plot(x, y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(x,y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='r') # No scatter plot, just a normal plot with marker

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10) 
ax.plot(x, y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(x,y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='r') 

ax.set_title('Anatomy of a figure', fontsize=20)

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10) 
ax.plot(x, y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(x,y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='r') 

ax.set_title('Anatomy of a figure', fontsize=20)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

ax.plot(x, y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10) 
ax.plot(x, y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(x,y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='r') 

ax.set_title('Anatomy of a figure', fontsize=20)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

ax.legend()

plt.show()
plt.close()
```
