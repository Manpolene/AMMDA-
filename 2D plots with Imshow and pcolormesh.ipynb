```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**2-D plots**\
There arde multiple ways to creat 2<D plots

**1. Imshow:** \
Imshow is short for image show and really shows the data as an image.

```python
x1 = np.linspace(0,5,50)
y1 = np.linspace(0,5,40)
z = np.sin(x1) ** 10 + np.cos(10 + y1[:, np.newaxis] * x1) * np.cos(x1)
```

```python
y1[:,np.newaxis]
```

```python
z
```

```python
z.shape
```

with this, we created a 2D array z based on x1 and y1 \
which can then be plotted e.g. using imshow

```python
fig = plt.figure(figsize=(5,3))

plt.imshow(z)

plt.show()
plt.close()
```

So we now can show the 2D array in a plot. If you look closer you will notice, that the (0,0) point, which we would expect to be in the lower left corner in a classic plot, is in the _upper_ left corner instead. \
This is because of handling the data as an image, not a mathamatical plot. \
So this kinda works but is not the solution to go for


**2. pcolormesh** \
Instead you will mostly use the pcolormesh to plot this kind of data. \
like imshow, we can give only the 2D array to the function.

```python
fig = plt.figure(figsize=(5,3))

plt.pcolormesh(z)

plt.show()
plt.close()
```

Now the 0,0 is as we want and that the plot is also flipped due to that 


We can also add the x and y data hier additionally:

```python
fig = plt.figure(figsize=(5,3))

plt.pcolormesh(x1, y1, z)

plt.show()
plt.close()
```

See the difference? \
While without giving informtion about the x and y values pcolormesh just uses equidistant points from 0 to the length of the data in each direction. \
If we also give x and y values, the plottting function is using these.

As we had for the scatterplots now it would also be nice to know what the colors are showing, for this we add a colorbar just as we did before:

```python
fig = plt.figure(figsize=(5,3))

plt.pcolormesh(x1, y1, z)
plt.colorbar()

plt.show()
plt.close()
```

**Editing colormap and limits** \
Like for scatter plots, we can also edit the cpolormap used and fix the limits of the colorbar

```python
fig = plt.figure(figsize=(5,3))

plt.pcolormesh(x1, y1, z, vmin = -1, vmax = 1, cmap = 'coolwarm')
plt.colorbar()

plt.show()
plt.close()
```

**Contour plots** \
A different way to plot this data would be to draw along the contours of the field.

```python
fig = plt.figure(figsize=(5,3))

plt.contour(x1, y1, z, levels = 20, cmap = 'coolwarm')

plt.show()
plt.close()
```

The arguemnt 'levels' gives how many different colored lines are plotted. Or how many levels the z data is splitted.

```python
fig = plt.figure(figsize=(5,3))

plt.contour(x1, y1, z, levels = 5, cmap = 'coolwarm')
plt.colorbar()

plt.show()
plt.close()
```

**Contourf (the contour filles)**

```python
fig = plt.figure(figsize=(5,3))

plt.contourf(x1, y1, z, levels = 5, cmap = 'coolwarm')
plt.colorbar()

plt.show()
plt.close()
```

```python
fig, axs = plt.subplots(1,3, sharey=True, figsize=(15,3))

p1 = axs[0].pcolormesh(x1, y1, z, vmin = -1.2, vmax = 1.2, cmap = 'coolwarm')
plt.colorbar(p1)
p2 = axs[1].contourf(x1, y1, z, levels = 5, cmap = 'coolwarm')
plt.colorbar(p2)
p3 = axs[2].contour(x1, y1, z, levels = 5, cmap = 'coolwarm')
plt.colorbar(p3)

plt.show()
plt.close()
```

See the differences between pcolormesh and contourf! \
In pcolormesh we plot the gridded data, contour draws a line (the contour) around the data, contourf just fills these contours

**MAp plotting** \
Map plotting is a nice skill to have in enviromental science, in many cases you want to show where your data was measured at or modelled from.

Especially together with Pandas or xarray, this is a powerful skill, but the basics we can learn without having to deal with pandas or xarray \
There are two main tools for creating maps, basemap and cartopy. In this course we will deal with cartopy.

```python
import cartopy.crs as ccrs
```

Back to figures and axes. While figures is the area where everything we draw lives on, axes is kind of the plotting space (or also the colorbar is a separate axes). \
Until now we only created a figure or created multiple axes when using a subplot. \
But we can also create an axes manully on an already created figure:

```python
fig = plt.figure(figsize=(5,3))
ax = plt.axes()
```

For maplotting we give matplotlib the information, that this axes now should be used for a map. We give a _projection_. \
In the first example we will work with the so called PlateCerree which looks like printed map. \
Also we want to draw the coastline directly onto this axes, note how we use ax.coastlines() and not plt.coastlines. \
Just with having the projection set to PlateCarree (or a different one, see later) we are able to plot the coastlines

```python
fig = plt.figure(figsize=(10,15))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(10,15))
ax = plt.axes(projection=ccrs.Mollweide())
ax.coastlines()

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(10,15))
ax = plt.axes(projection=ccrs.Robinson())
ax.coastlines()

plt.show()
plt.close()
```

**Labels** \
Until now we just have a map plot, but if we plot data gridlines and especially ticks on the x and y axis would be necessary \
Here a lot happend in the last versions, in the current version it is kind of easy to include everything:

```python
plt.figure(figsize=(15,5))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.OCEAN)

bremen_lon=8.8016937
bremen_lat=53.0792962
toronto_lon=-79.383087
toronto_lat=43.66135

plt.plot([bremen_lon,toronto_lon],[bremen_lat,toronto_lat],color='red',transform=ccrs.Geodetic())

ax.set_global()


gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, 
                  linewidth=2, color='green', alpha=0.2, linestyle='--')

plt.show()
plt.close()
```

We can turn some of the gridline labels off if we don't want to have them on every side:

```python
import matplotlib.ticker as mticker
```

```python
plt.figure(figsize=(15,5))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.OCEAN)

bremen_lon=8.8016937
bremen_lat=53.0792962
toronto_lon=-79.383087
toronto_lat=43.66135

plt.plot([bremen_lon,toronto_lon],[bremen_lat,toronto_lat],color='red',transform=ccrs.Geodetic())

ax.set_extent([-180, 180, -90, 90])


gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=2, color='green', alpha=0.2, linestyle='--')
gl.top_labels=False
gl.left_labels=False

gl.xlocator = mticker.FixedLocator(np.arange(-180,181,5))

plt.show()
plt.close()
```
